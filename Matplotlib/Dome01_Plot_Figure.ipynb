```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
'''
Author: Szl
Date: 2024-03-13 15:12:47
LastEditors: Szl
LastEditTime: 2024-03-13 15:30:22
Description: 
'''
# 1.创建画布，指定尺寸为5 * 3
plt.Figure(figsize=(5, 3))

# 2.准备输入x和输出y，并同时构造x与y之间的函数映射关系
x = np.linspace(0, 2 * np.pi) # 给定x轴的取值，这个取值同时指定了x的取值范围，从0-2Π
y = np.sin(x)

# 绘制sin函数的图像
plt.plot(x, y)

# 第二个新的函数映射
m = np.cos(x)
plt.plot(x, m)

# 绘制网格线
plt.grid(
  linestyle = ':',
  color = 'red',
  alpha = 0.5
)

# 设置坐标轴范围
plt.axis([-1, 7, -1.5, 1.5])

# 保存图片
plt.savefig(
  './img1.png',
  dpi = 200,
  facecolor = 'white',
  edgecolor = 'red',
  bbox_inches = 'tight'
)
```

```python
# 创建画布
plt.Figure(figsize = (10, 6))

# 给定x和y以及映射关系
x = np.linspace(0, 2 * np.pi, 20)
y1 = np.sin(x) # 正弦函数
y2 = np.cos(x) # 余弦函数

# 设置曲线的颜色 线型 点型
plt.plot(x, y1, color = 'green', ls = '-.', marker = 'p')
plt.plot(x, y2, color = 'indigo', ls = '--', marker = 'o')
plt.plot(x, y1 + y2, color = 'blue', ls = ':', marker = '*')
plt.plot(x, y1 + 2 * y2, color = 'orange', linewidth = 3, alpha = 0.3)

# 保存图片
plt.savefig(
  './img2.png',
  dpi = 200,
  facecolor = 'white',
  edgecolor = 'red',
  bbox_inches = 'tight'
)
```

```python
plt.figure(figsize = (10, 6))

def f(x):
  return np.exp(-x) * np.cos(2 * np.pi * x)

x = np.linspace(0, 5, 50)

plt.plot(
  x,
  f(x),
  marker = 'o',
  color = 'indigo',
  ls = '--',
  lw = 2,
  alpha = 0.5,
  markerfacecolor = 'red',
  markeredgecolor = 'green',
  markeredgewidth = 3
)

# 设置刻度大小
plt.xticks(size = 18)
plt.yticks(size = 18)

plt.savefig(
  './img3.png',
  dpi = 200,
  facecolor = 'white',
  edgecolor = 'red',
  bbox_inches = 'tight'
)
```

```python
plt.Figure(figsize = (10, 6))
x = np.linspace(-np.pi, np.pi, 50)
y = np.sin(x)

# 子视图区域的绘制
# 整个绘图区域基本按照2 * 2形状进行布局，即子视图区域是2行2列

#子视图区域1
ax = plt.subplot(2, 2, 1) # 在2行2列的区域中的第一个子视图下绘制图像
ax.plot(x, y, color = 'red')

# 设置子视图区域的整体属性
ax.set_facecolor('green')

# 子视图区域2
ax =  plt.subplot(2, 2, 2) # 在2行2列的区域中的第二个子视图下绘制图像
line, = ax.plot(x, -y)
line.set_marker('*')
line.set_markerfacecolor('red')
line.set_markeredgecolor('green')
line.set_markersize(3)

# 子视图区域3
ax = plt.subplot(2, 1, 2)
plt.sca(ax) # 设置轴为当前视图 - 轴就是坐标轴，即在当前子视图区域的图像展示坐标轴
x = np.linspace(-np.pi, np.pi, 200)
plt.plot(x, np.sin(x * x), color = 'red')

plt.savefig(
  './img4.png',
  dpi = 200,
  facecolor = 'white',
  edgecolor = 'red',
  bbox_inches = 'tight'
)
```

```python
fig = plt.Figure(figsize = (10, 6))

x = np.linspace(-np.pi, np.pi, 25)
y = np.cos(x)

plt.plot(x, y)

# 在大图画布下嵌套小的子视图图像
ax = plt.axes([0.2, 0.55, 0.3, 0.3])
ax.plot(x, y, color = 'green')

# 使用大视图对象 - fig添加子视图的方式嵌套一个子视图
ax = fig.add_axes([0.5, 0.15, 0.3, 0.3])
ax.plot(x, y, color = 'red')
```

```python
fig,((ax11, ax12, ax13), (ax21, ax22, ax23), (ax31, ax32, ax33)) = plt.subplots(3, 3)
# 返回值的含义：
  # fig：是返回的整个大的画布
  #元组内的三行三列对象表示符合三行三列布局的9个子视图区域的绘图对象

fig.set_figwidth(10)
fig.set_figheight(6)

x = np.linspace(0, 2 * np.pi)

ax11.plot(x, np.sin(x))
ax12.plot(x, np.cos(x))
ax13.plot(x, np.tan(x))
ax21.plot(x, np.tanh(x))
ax22.plot(x, np.arcsin(x))
ax23.plot(x, np.arccos(x))
ax31.plot(x, np.arctan(x))
ax32.plot(x, np.sin(x) + np.cos(x))
ax33.plot(x, np.sin(x) + np.tan(x))

plt.tight_layout()

plt.savefig(
  './img5.png',
  dpi = 200,
  facecolor = 'white',
  edgecolor = 'red',
  bbox_inches = 'tight'
)
```

```python
fig = plt.figure(figsize = (10, 6))

x = np.linspace(-np.pi, np.pi, 100)
y1 = np.exp(x)
y2 = np.sin(x)

# 可以设置画布区域中整体的文字字体大小
plt.rcParams['font.size'] = 18

# 获取当前的轴域
ax1 = plt.gca()

ax1.set_xlabel('time(s)')
ax1.set_ylabel('exp', color = 'red')

ax1.plot(x, y1, color = 'green')

# 坐标轴的属性
ax1.tick_params(axis = 'y', labelcolor = 'blue')

# 在当前轴域下创建一个新的axes实例
ax2 = ax1.twinx()
ax2.set_ylabel('sin', color = 'yellow')

ax2.plot(x, y2, color = 'orange')
ax1.tick_params(axis = 'y', labelcolor = 'black')
plt.tight_layout() # 设置紧凑布局

plt.savefig(
  './img6.png',
  dpi = 200,
  facecolor = 'white',
  edgecolor = 'red',
  bbox_inches = 'tight'
)
```

```python
# 堆叠柱状图绘制

# 给一组横坐标的label
labels = ['G1', 'G2', 'G3', 'G4', 'G5', 'G6']

# 两组数据
men_mean = np.random.randint(20, 35, size = 6)
women_mean = np.random.randint(20, 35, size = 6)

men_std = np.random.randint(1, 7, size = 6)
women_std = np.random.randint(1, 7, size = 6)

# 柱子的宽度
width = 0.35

# 绘图
plt.bar(
  labels,
  men_mean,
  width,
  yerr = 4,
  label = 'Men',
)
plt.bar(
  labels,
  women_mean,
  width,
  yerr = 2,
  bottom = men_mean,
  label = 'Women',
)

# 设置一个y轴标签
plt.ylabel('Scores')

# 标题
plt.title('Scores by group and gender')

# 设置一个图例
plt.legend()

plt.savefig(
  './img7.png',
  dpi = 200,
  facecolor = 'white',
  edgecolor = 'red',
  bbox_inches = 'tight'
)
```

```python
x = np.arange(len(men_mean))

plt.figure(figsize = (10, 6))

resc1 = plt.bar(
  x - width / 2,
  men_mean,
  width
)
resc2 = plt.bar(
  x + width / 2,
  women_mean,
  width
)

# 设置标题图例
plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(x, labels)
plt.legend(['Men','Women'])

# 添加注释
def set_label(rescs):
  for resc in rescs:
    height = resc.get_height()
    plt.text(
      x = resc.get_x() + resc.get_width() / 2, # 水平坐标
      y = height + 0.5, # 竖直坐标
      s = height,
      ha = 'center' # 水平居中
    )
    
set_label(resc1)
set_label(resc2)
plt.tight_layout()    
```

```python
# 构造一组fake data

# 均值
mu = 100 # μ

# 标准差
sigma = 15 # σ

x = np.random.normal(loc = mu, scale = 15, size = 10000)

# 创建绘图对象
fix,ax = plt.subplots()

# 直方图绘制
n, bins, patchs = ax.hist(x, 200, density = True, facecolor = 'g', alpha = 0.75)
# 注意：hist方法只会绘制直方图，但是一般情况下，对样本进行绘制所谓的直方图时，一般都会带上其随机变量的概率密度函数
# 直方图 + 概率密度函数的曲线，二者组合在一起的增图示，叫做“高斯核密度曲线”

# 这是一个符合正态分布的随机变量的概率密度函数
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) * np.exp(-0.5 * (1 / sigma * (bins - mu)) ** 2))

# 绘制概率密度函数的曲线 - 高斯核密度曲线
plt.plot(bins, y, '--')

plt.xlabel('Smarts')
plt.ylabel('Probability density')
plt.title(r'Histogram of IQ: $\mu = 100$, $\sigma = 15$')
fig.tight_layout()

plt.savefig(
  './img8.png',
  dpi = 200,
  facecolor = 'white',
  edgecolor = 'red',
  bbox_inches = 'tight'
)
```
