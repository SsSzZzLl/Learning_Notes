```python
import numpy as np
import pandas as pd
```

```python
# 1.使用list创建一个dataframe
data = [
  ['szl', 20],
  ['SSSzll', 23],
  ['song', 21]
]
data1 = pd.DataFrame(data = data, columns = ['name', 'age'], index = ['A', 'B', 'C'], dtype = 'object')
data1
```

```python
#使用ndarray创建一个dataframe
df = pd.DataFrame(data = np.random.randint(60, 100, size = (3, 3)), dtype = 'int32', columns = ['Python', 'Java', 'Pytorch'], index = ['szl', 'SSSzll', 'song'])
df
```

```python
#使用一个字典构建一个dataframe对象
'''
注意：如果使用字典来构建dataframe对象，则字典对象的内部结构具有一个统一化的格式：
data = {
  '键名[列索引名称]' : 键值一般给定一个list或tuple甚至一个Series ———— 当前这个完整的键值对就表示：一个名叫XXX的列取到了那些值
}
'''

data = {
  'Python' : [34, 78, 98],
  'Pytorch' : [45, 76, 87],
  'Java' : [78, 87, 93]
}

df = pd.DataFrame(
  data = data,
  index = ['szl', 'SSSzll', 'song'],
  dtype = 'int32'
)
df
```

```python
df = pd.DataFrame(
  data = np.random.randint(0, 10, size = (10, 3)),
  index = list('ABCDEFGHIJ'),
  columns = ['seaborn', 'pandas', 'Numpy']
)
df
```

```python
#map的第一种使用方法,直接使用字典来映射值
df['Numpy'].map(
  {
    1 : 'hello',
    2 : 'world',
    3 : 'python'
  }
)
```

```python
# map的第二种使用方式，使用自定义函数
def conv(x):
  if x % 3 == 0:
    return True
  else:
    return False

df['pandas'].map(conv)
```
